```python
# libraries
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
from geopy.geocoders import Nominatim
```

```python
m = Basemap(projection='cyl',llcrnrlat=-80,urcrnrlat=80,llcrnrlon=-180,urcrnrlon=180,resolution='c', lon_0=50)
m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m.fillcontinents(color='grey', alpha=0.7, lake_color='lightblue')
m.drawcoastlines(linewidth=0.1, color="white")
m.drawcountries(color="lightblue")

# Draw one travel
geolocator = Nominatim()
loc1 = geolocator.geocode("Beijing")
loc2= geolocator.geocode("Copenhagen")
m.drawgreatcircle(loc1.longitude, loc1.latitude, loc2.longitude, loc2.latitude, linewidth=1, color='orange')

# Draw another travel
geolocator = Nominatim()
loc3 = geolocator.geocode("Billund")
loc4= geolocator.geocode("New York")
m.drawgreatcircle(loc3.longitude, loc3.latitude, loc4.longitude, loc4.latitude, linewidth=1, color='red')

plt.show(m)
```
